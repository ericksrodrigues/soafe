```python
diabetes_df = pd.read_csv("./datasets/diabetes.csv")
hill_valley_df = pd.read_csv("./datasets/hill_valley.csv")
ionosphere_df = pd.read_csv("./datasets/ionosphere.csv")
megawatt1_df = pd.read_csv("./datasets/megawatt1.csv")
sonar_df = pd.read_csv("./datasets/sonar.csv")
spam_df = pd.read_csv("./datasets/spam.csv")
```

```python
def statistics(df):
    linhas, colunas = df.shape
    print("Número de Registros:", linhas)
    print("Número de Features:", colunas-1)
    print("Amostras:")
    print(df.sample(5))
    print("Descrição:")
    print(df.describe())
    
 
```
